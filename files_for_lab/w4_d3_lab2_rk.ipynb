1. Get the numerical variables from our dataset.
2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')

numericals = data.select_dtypes('number')
numericals.head()
numericals.shape[1]
```

```python
def plot_numericals(df):
    for col in df.columns:
        plt.figure()
        sns.histplot(x=df[col])

plot_numericals(numericals)
```

```python
def plot_numericals(df):
    count= df.shape[1]
    fig, ax = plt.subplots(nrows=1,ncols=count,figsize=(25,5))
    for i,col in enumerate(df.columns):
        sns.histplot(x=df[col], ax=ax[i])
    plt.show()

plot_numericals(numericals)
```

```python
# None of the data fits to a normal or exponential distribution
```

```python
# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

from sklearn.preprocessing import MinMaxScaler, StandardScaler, PowerTransformer

standardizer = StandardScaler()
standardizer.fit(numericals)

minmaxizer = MinMaxScaler()
minmaxizer.fit(numericals)

standard = standardizer.transform(numericals)
standard = pd.DataFrame(standard, columns=numericals.columns)

minmax = minmaxizer.transform(numericals)
minmax = pd.DataFrame(minmax, columns=numericals.columns)

powerizer = PowerTransformer()
powerizer.fit(numericals)

power = powerizer.transform(numericals)
power = pd.DataFrame(power, columns=numericals.columns)


def logizer(df):
    local=df.copy()
    for x in local.columns:
        local[x]=np.log(local[x])
    return local  

log = logizer(numericals)
```

```python
plot_numericals(standard)
```

```python
plot_numericals(minmax)
```

```python
plot_numericals(power)
```

```python
plot_numericals(log)
```
